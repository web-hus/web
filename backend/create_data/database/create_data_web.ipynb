```python
# Chuyển thành tuple để tối ưu bộ nhớ và tốc độ truy xuất
LAST_NAMES: Tuple[str, ...] = ("Nguyễn", "Trần", "Lê", "Phạm", "Hoàng", "Huỳnh", "Phan", "Vũ", "Võ", "Đặng", "Bùi", "Đỗ", "Hồ", "Ngô", "Dương", "Lý")
MIDDLE_NAMES_MALE: Tuple[str, ...] = ("Anh", "Minh", "Văn", "Gia", "Quang", "Thiên")
MIDDLE_NAMES_FEMALE: Tuple[str, ...] = ("Thị", "Thu", "Tuệ", "Thùy", "Khánh", "An", "Bích", "Xuân")
FIRST_NAMES_MALE: Tuple[str, ...] = ("Huy", "Khang", "Bảo", "Minh", "Phúc", "Khoa", "Thuận", "Tuấn", "Tú", "Tùng", "Nam", "Tiến", "Phát", "Lộc", "Minh", "Hồ", "Trung", "Bách")
FIRST_NAMES_FEMALE: Tuple[str, ...] = ("Linh", "Chi", "Thư", "Nguyệt", "Lan", "Như", "Trinh", "Thảo", "Trang", "Hằng", "Nhung", "Vân", "Ly", "Huyền", "Đào", "Thúy")

def gen_random_password(name: str) -> str:
    random_numbers = ''.join(random.choices(string.digits, k=3))
    special_char = random.choice("!@#$%^&*()-_=+[]{}|;:,.<>?/")
    password = f"{name}{random_numbers}{special_char}"
    
    return password

with open(r'C:\Users\Admin\Desktop\address.txt', 'r', encoding='utf8') as file:
    addresses = [line.strip() for line in file]
address2 = list(set(addresses))

generated_phone_numbers = set()

def gen_random_phone_number() -> str:
    # Sinh số điện thoại ngẫu nhiên
    phone_number = f"0{random.randint(1, 9)}{''.join(str(random.randint(0, 9)) for _ in range(8))}"
    
    # Đảm bảo số điện thoại duy nhất
    while phone_number in generated_phone_numbers:
        phone_number = f"0{random.randint(1, 9)}{''.join(str(random.randint(0, 9)) for _ in range(8))}"
    
    # Thêm số điện thoại vào bộ để theo dõi
    generated_phone_numbers.add(phone_number)
    
    return phone_number

generated_emails = set()

def gen_random_email(gender, name) -> str:
    while True:
        # Sinh 5 số ngẫu nhiên
        random_numbers = ''.join(random.choices(string.digits, k=5))

        # Tạo email theo định dạng: tên + họ + 5 số ngẫu nhiên + @gmail.com
        email = f"{name}{random_numbers}@gmail.com"

        # Kiểm tra email có duy nhất không
        if email not in generated_emails:
            generated_emails.add(email)  # Thêm vào tập hợp nếu chưa tồn tại
            return email

def create_users_info(n: int) -> pd.DataFrame:
    """Tạo DataFrame với thông tin người dùng"""
    
    genders = np.random.choice(['M', 'F'], size=n)
    ages = np.random.randint(18, 66, n)
    user_roles = np.zeros(n, dtype=int)
    
    
    names = []
    passwords = []
    phones = []
    email = []
    address = []
    
    for gender in genders:
        if gender == 'F':
            last = random.choice(LAST_NAMES)
            mid = random.choice(MIDDLE_NAMES_FEMALE)
            first = random.choice(FIRST_NAMES_FEMALE)

            name = f"{last} {mid} {first}"
            name_email = f"{first} {last} {mid}"
            name_pw = f"{last} {first}"

            names.append(name)
            passwords.append(gen_random_password(unidecode(name_pw).lower().replace(" ", "")))
            email.append(gen_random_email(gender, unidecode(name_email).lower().replace(" ", "")))
        else:  # 'M'
            last = random.choice(LAST_NAMES)
            mid = random.choice(MIDDLE_NAMES_MALE)
            first = random.choice(FIRST_NAMES_MALE)

            name = f"{last} {mid} {first}"
            name_email = f"{first} {last} {mid}"
            name_pw = f"{last} {first}"

            names.append(name)
            passwords.append(gen_random_password(unidecode(name_pw).lower().replace(" ", "")))
            email.append(gen_random_email(gender, unidecode(name_email).lower().replace(" ", "")))
        add = random.choice(address2)
        address.append(add)
        phones.append(gen_random_phone_number())
    
    users_df = pd.DataFrame({
        'user_name': names,
        'gender': genders,
        'age': ages,
        'user_password': passwords,
        'user_role': user_roles,
        'phone': phones,
        'email': email,
        'address':address
    })
    
    
    return users_df


users = create_users_info(1000)
```

```python
import random
from datetime import datetime, timedelta

# Set seed for reproducibility
random.seed(42)

# Define time intervals
morning_start = '10:00'
morning_end = '14:00'
evening_start = '18:00'
evening_end = '22:00'

# Function to generate time slots in 'HH:MM:SS' format
def generate_time_slots(start_time, end_time):
    start = datetime.strptime(start_time, '%H:%M')
    end = datetime.strptime(end_time, '%H:%M')
    time_slots = []
    
    # Generate time slots with 30-minute intervals
    while start <= end:
        time_slots.append(start.strftime('%H:%M:%S'))
        start += timedelta(minutes=30)
    
    return time_slots

# Generate all possible time slots for morning and evening
morning_slots = generate_time_slots(morning_start, morning_end)
evening_slots = generate_time_slots(evening_start, evening_end)

# Combine both morning and evening slots
all_slots = morning_slots + evening_slots

# Generate 200 random records, allowing for duplicates
time = [random.choice(all_slots) for _ in range(200)]

# Print the list of 200 records
for record in time:
    print(record)
```

```python
import random
from datetime import datetime, timedelta

# Set seed for reproducibility
random.seed(42)

# Define the date range from October 2024 to January 2025
start_date = datetime(2024, 10, 1)
end_date = datetime(2025, 1, 31)
cutoff_date = datetime(2025, 1, 10)  # Date cutoff: 10/1/2025

# Function to generate random dates within the given range
def generate_random_dates(start_date, end_date, num_dates):
    delta_days = (end_date - start_date).days
    random_dates = []
    
    for _ in range(num_dates):
        random_days = random.randint(0, delta_days)
        random_date = start_date + timedelta(days=random_days)
        random_dates.append(random_date.strftime('%Y-%m-%d'))  # Format as 'YYYY-MM-DD'
    
    return random_dates

# Generate 200 random dates
random_dates = generate_random_dates(start_date, end_date, 200)

# Generate the statuses list
statuses = []
for date in random_dates:
    date_obj = datetime.strptime(date, '%Y-%m-%d')
    if date_obj <= cutoff_date:
        # 90% are 1 (active), 10% are 0 (cancelled)
        status = random.choices([1, 2], weights=[0.9, 0.1])[0]
    else:
        # After 10/1/2025, status is 2 (waiting)
        status = 0
    statuses.append({"date": date, "status": status})

# Print the result
for entry in statuses:
    print(entry)
```

```python
import random

# Đặt seed để tạo dữ liệu ngẫu nhiên có thể tái tạo
random.seed(42)

# Giả định bạn đã có danh sách order_id và dish_id
order_ids = [i for i in range(1, 451)]  # Tạo danh sách order_id từ 1 đến 450
dish_ids = [f'D{i:03}' for i in range(1, 68)]  # Tạo danh sách dish_id từ D001 đến D067

# Danh sách để lưu các câu lệnh SQL
sql_statements = []

# Tạo 450 câu lệnh INSERT
for order_id in order_ids:
    # Chọn ít nhất một dish_id ngẫu nhiên cho mỗi order_id
    num_dishes = random.randint(1, 3)  # Mỗi order_id có từ 1 đến 3 dish_id
    chosen_dishes = random.sample(dish_ids, num_dishes)  # Chọn các dish_id ngẫu nhiên
    
    # Thêm các câu lệnh vào sql_statements
    for dish_id in chosen_dishes:
        quantity = random.randint(1, 3)  # Số lượng ngẫu nhiên từ 1 đến 3
        sql_statements.append(f"({order_id}, '{dish_id}', {quantity})")

# In ra tất cả câu lệnh SQL, mỗi câu lệnh sẽ được cách nhau bằng dấu phẩy
print(",\n".join(sql_statements))
```
